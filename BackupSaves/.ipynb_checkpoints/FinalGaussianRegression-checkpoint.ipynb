# Gaussian Process Regression

```python
import scipy.io
import numpy 
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
data_xvals = numpy.atleast_2d( [ 1.0, 3.0, 5.0, 6.0, 7.0, 8.0 ] )
data_yvals = numpy.sin( data_xvals )
plt.scatter(data_xvals,data_yvals,5)
plt.plot(data_xvals,data_yvals)
plt.show()
```


## Linear - No Noise


```python
testValues = numpy.linspace(-2,10,200)
output = []
variance =[]
for x in testValues:
    A = x*data_xvals
    B = numpy.linalg.pinv(numpy.dot(data_xvals.T,data_xvals)) # x,x
    C = data_yvals.T
    D = numpy.dot(A,B)
    E = numpy.dot(D,C)
    output.append(E)
    A1 = x*x
    B1 = x*data_xvals
    C1 = numpy.linalg.pinv(numpy.dot(data_xvals.T,data_xvals))
    D1 = (data_xvals*x).T
    E1 = numpy.dot(B1,C1)
    F1 = numpy.dot(E1,D1)
    variance.append(A1 - F1)

y1 = [(x+y)[0][0] for x,y in zip(output,variance)] 
y2 = [(x-y)[0][0] for x,y in zip(output,variance)]

plt.scatter(data_xvals,data_yvals,5)
plt.plot(data_xvals,data_yvals)
plt.fill_between(testValues,y1,y2,alpha=0.5)
plt.scatter(testValues,output,2)
plt.show()
```

## Linear Noise

```python
testValues = numpy.linspace(-2,10,100)
output = []
variance =[]
for x in testValues:
    A = x*data_xvals # xtest,x
    B = numpy.linalg.pinv(numpy.dot(data_xvals.T,data_xvals) + (numpy.identity(6)*.1)) # x,x
    C = data_yvals.T
    D = numpy.dot(A,B)
    E = numpy.dot(D,C)
    output.append(E)
    A1 = x*x
    B1 = x*data_xvals
    C1 = numpy.linalg.pinv(numpy.dot(data_xvals.T,data_xvals) + (numpy.identity(6)*.1))
    D1 = (data_xvals*x).T
    E1 = numpy.dot(B1,C1)
    F1 = numpy.dot(E1,D1)
    variance.append(A1 - F1)

y1 = [(x+y)[0][0] for x,y in zip(output,variance)] 
y2 = [(x-y)[0][0] for x,y in zip(output,variance)]


plt.scatter(data_xvals,data_yvals,5)
plt.plot(data_xvals,data_yvals)
plt.fill_between(testValues,y1,y2,alpha=0.5)
plt.scatter(testValues,output,2)
plt.show()
```

## Polynomial - No Noise

```python
output = []
variance = []
for x in testValues:
    A = numpy.power(1+x*data_xvals,3) # xtest,x
    B = numpy.power(1+numpy.dot(data_xvals.T,data_xvals),3)# x,x
    C = data_yvals.T
    noisyB = numpy.linalg.pinv(B) # noise
    D = numpy.dot(A,noisyB)
    E = numpy.dot(D,C)   
    output.append(E[0])
    A1 = numpy.power(1+x*x,3)
    B1 = numpy.power(1+x*data_xvals,3)
    C1 = numpy.linalg.pinv(numpy.power(1+numpy.dot(data_xvals.T,data_xvals),3)) #noise
    D1 = (numpy.power(1+data_xvals*x,3).T)
    E1 = numpy.dot(B1,C1)
    F1 = numpy.dot(E1,D1)
    G1 = A1 - F1
    variance.append(G1[0])   

y1 = [(x+y)[0] for x,y in zip(output,variance)] 
y2 = [(x-y)[0] for x,y in zip(output,variance)]

plt.scatter(data_xvals,data_yvals,5)
plt.plot(testValues,output)
plt.fill_between(testValues,y1,y2, alpha=0.5)
```

## Polynomial Noise

```python
output = []
variance = []
for x in testValues:
    A = numpy.power(1+x*data_xvals,3) # xtest,x
    B = numpy.power(1+numpy.dot(data_xvals.T,data_xvals),3)# x,x
    C = data_yvals.T
    noisyB = numpy.linalg.pinv(B + (numpy.identity(6)*.1)) # noise
    D = numpy.dot(A,noisyB)
    E = numpy.dot(D,C)   
    output.append(E[0])
    A1 = numpy.power(1+x*x,3)
    B1 = numpy.power(1+x*data_xvals,3)
    C1 = numpy.linalg.pinv(numpy.power(1+numpy.dot(data_xvals.T,data_xvals),3) + (numpy.identity(6)*.1)) #noise
    D1 = (numpy.power(1+data_xvals*x,3).T)
    E1 = numpy.dot(B1,C1)
    F1 = numpy.dot(E1,D1)
    G1 = A1 - F1
    variance.append(G1[0])   
y1 = [(x+y)[0] for x,y in zip(output,variance)] 
y2 = [(x-y)[0] for x,y in zip(output,variance)]

plt.scatter(data_xvals,data_yvals,5)
plt.plot(testValues,output)
plt.fill_between(testValues,y1,y2, alpha=0.5)
```

## RBF - No Noise

```python
output = []
variance = []
for x in testValues:
    A = numpy.exp((-.5)*numpy.power(x-data_xvals,2))
    B = numpy.linalg.pinv(numpy.exp((-.5)*numpy.power(data_xvals.T-data_xvals,2))) # no Noise
    C = data_yvals.T
    D = numpy.dot(A,B)
    E = numpy.dot(D,C)
    output.append(E[0])
    A1 = numpy.exp((-.5)*numpy.power(x-x,2))
    B1 = numpy.exp((-.5)*numpy.power(x-data_xvals,2))
    C1 = numpy.linalg.pinv(numpy.exp((-.5)*numpy.power(data_xvals.T-data_xvals,2)))
    D1 = numpy.exp((-.5)*numpy.power(data_xvals-x,2)).T
    E1 = numpy.dot(B1,C1)
    F1 = numpy.dot(E1,D1)
    G1 = A1 - F1
    variance.append(G1[0])

y1 = [(x+y)[0] for x,y in zip(output,variance)] 
y2 = [(x-y)[0] for x,y in zip(output,variance)]

plt.scatter(data_xvals,data_yvals,5)
plt.fill_between(testValues,y1,y2,alpha=0.5)    
plt.plot(testValues,output)
plt.show()
```

## RBF Noise

```python
output = []
variance = []
for x in testValues:
    A = numpy.exp((-.5)*numpy.power(x-data_xvals,2))
    B = numpy.linalg.pinv(numpy.exp((-.5)*numpy.power(data_xvals.T-data_xvals,2)) + (numpy.identity(6)*.1)) # no Noise
    C = data_yvals.T
    D = numpy.dot(A,B)
    E = numpy.dot(D,C)
    output.append(E[0])
    A1 = numpy.exp((-.5)*numpy.power(x-x,2))
    B1 = numpy.exp((-.5)*numpy.power(x-data_xvals,2))
    C1 = numpy.linalg.pinv(numpy.exp((-.5)*numpy.power(data_xvals.T-data_xvals,2)) + (numpy.identity(6)*.1))
    D1 = numpy.exp((-.5)*numpy.power(data_xvals-x,2)).T
    E1 = numpy.dot(B1,C1)
    F1 = numpy.dot(E1,D1)
    G1 = A1 - F1
    variance.append(G1[0])

y1 = [(x+y)[0] for x,y in zip(output,variance)] 
y2 = [(x-y)[0] for x,y in zip(output,variance)]

plt.scatter(data_xvals,data_yvals,5)
plt.fill_between(testValues,y1,y2, alpha=0.5)    
plt.plot(testValues,output)
plt.show()
```

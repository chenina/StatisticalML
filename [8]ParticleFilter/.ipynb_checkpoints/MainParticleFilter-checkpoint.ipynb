```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn

def create_map():
    room_map = np.asarray([

        # outer walls
        [0, 0, 0, 1], 
        [0, 1, 1, 1],  
        [1, 1, 1, 0],  
        [1, 0, 0, 0],  

        # upper rooms
        [0,    0.6, 0.15, 0.6],  
        [0.25, 0.6, 0.6,  0.6],  
        [0.7,  0.6, 1.0,  0.6],
        [0.4,  1.0, 0.4,  0.6],

        # lower rooms
        [0,    0.4, 0.15, 0.4],  
        [0.25, 0.4, 0.8,  0.4],  
        [0.9,  0.4, 1.0,  0.4],
        [0.4,  0.0, 0.4,  0.4],

        # "table"
        [0.1, 0.1, 0.1, 0.15],  
        [0.1, 0.15, 0.15, 0.15],  
        [0.15, 0.15, 0.15, 0.1]
    ])

    return room_map

# adapted from the original matlab code at
# http://www.mathworks.com/matlabcentral/fileexchange/27205-fast-line-segment-intersection
def line_intersect( X1,Y1,X2,Y2, X3,Y3,X4,Y4 ):

    X4_X3 = X4.T - X3.T
    Y1_Y3 = Y1   - Y3.T
    Y4_Y3 = Y4.T - Y3.T
    X1_X3 = X1   - X3.T
    X2_X1 = X2   - X1
    Y2_Y1 = Y2   - Y1

    numerator_a = X4_X3 * Y1_Y3 - Y4_Y3 * X1_X3
    numerator_b = X2_X1 * Y1_Y3 - Y2_Y1 * X1_X3
    denominator = Y4_Y3 * X2_X1 - X4_X3 * Y2_Y1

    u_a = numerator_a / denominator
    u_b = numerator_b / denominator 

    INT_X = X1 + X2_X1 * u_a
    INT_Y = Y1 + Y2_Y1 * u_a
    INT_B = (u_a >= 0) & (u_a <= 1) & (u_b >= 0) & (u_b <= 1)

    return INT_X, INT_Y, INT_B

def show_map( room_map ):
    for i in range(0,room_map.shape[0]):
        plt.plot( [room_map[i,0], room_map[i,2]], [room_map[i,1], room_map[i,3]] )

    plt.axis('equal')
    plt.xlim([-0.1, 1.1])
    plt.ylim([-0.1, 1.1])    

def vec_linspace( a, b, num=10 ):
    return np.asmatrix( [
        np.linspace( a[0,0], b[0,0], num=num ),
        np.linspace( a[0,1], b[0,1], num=num ),
        np.linspace( a[0,2], b[0,2], num=num )        
    ] )

def cast_rays( parts, room_map, verbose=False ):
    sensor_thetas = 0.1 * np.asarray( range(-5,6) )

    px = parts[0:1,:].T
    py = parts[1:2,:].T
    pt = parts[2:3,:].T

    rx1 = room_map[:,0:1]
    ry1 = room_map[:,1:2]
    rx2 = room_map[:,2:3]
    ry2 = room_map[:,3:4]

    dl = []
    for theta in sensor_thetas:
        # calculate intersections between all rays and all map lines
        int_x, int_y, int_b = line_intersect( px, py,
                                              px + np.sin(theta+pt),
                                              py + np.cos(theta+pt),
                                              rx1, ry1, rx2, ry2 )

        inds = int_b.ravel()

        # calculate nearest intersections
        # simplify code by setting non-intersections to be far away
        n_int_b = np.logical_not( int_b )
        int_x[ n_int_b ] = 1000 
        int_y[ n_int_b ] = 1000

        # calculate distances
        dx = px-int_x
        dy = py-int_y
        dists = dx*dx + dy*dy
        min_dists = np.min( dists, axis=1, keepdims=True )
        dl.append( min_dists )

        # show rays from a point, and their intersections
        # only needed for visualization
        if verbose:
            min_pts = np.argmin( dists, axis=1 )
            for i in range( 0, px.shape[0] ):
                plt.plot( [ px[i,0], int_x[i,min_pts[i]] ],
                          [ py[i,0], int_y[i,min_pts[i]] ] )

    data = np.hstack( dl )

    return data
```

```python
#Data Initialization
matlabData = scipy.io.loadmat('sensors.mat')
sonarsData = matlabData['sonars']
trueStates = matlabData['true_states']
print trueStates.shape # x , y , directional angle facing 
print sonarsData.shape # 11 lazers reading distance
```

```python
# TIE IT ALL TOGETHER
particles = []
predictedXs = []
predictedYs = []

allGaussianWeights = []
allLaplacianWeights = []
allPrevoiusWeights = []

# ==========================   CREATE STARTING DISTRIBUTION ======================= ##

#xSpread = 10
#ySpread = 10
#n = xSpread * ySpread
n = lenth
distribution = np.zeros((3,n), dtype=np.float)

#for i in range(0,n,1):
#    distribution[0,i] = random.uniform(0, 6.28)
#    distribution[1,i] = random.uniform(0, 6.28)
#    distribution[2,i] = random.uniform(0, 6.28)
    
counter = 0
for i in np.linspace(0, 1, num=xSpread, dtype=np.float):
    
    for k in np.linspace(0, 1, num=ySpread, dtype=np.float):
        print 'k', k
        print 'i ', i
        distribution[0,counter] = i
        distribution[1,counter] = k
        distribution[2,counter] = random.uniform(0, 6.28) #between 0 and 2pi
        counter += 1

raysCastFromRobot =  sonarsData #cast_rays(trueStates, create_map()).T # Replace with sonars
particles.append(distribution)
# ================================================================================ ##


# Itterate through all time steps of robot movement
for t in range (0,len(sonarsData[0]),1): 
    raysT = raysCastFromRobot[:,t] # use sonar data
    raysCastFromRobotAtDistributionPoints = cast_rays(particles[t], create_map()).T
   
    #Create Gaussian Weights for each particle
    gaussianWeights = np.zeros((n), dtype=np.float);
    laplacianWeights = np.zeros((n), dtype=np.float);
    
    for distributionIndex in range (0,len(particles[t][0]),1):
        difference = raysT - raysCastFromRobotAtDistributionPoints[:,distributionIndex]
        gaussLikelihood = math.exp((-1/2)* difference.dot(difference.T)) # * NO prior on first step
        laplacianLikelihood = math.exp((-1/2)* sum(abs(difference)))
        prior = 1
        if t == 0: 
            prior = 1
        else: 
            prior = allPrevoiusWeights[t-1][distributionIndex]
            #print prior
        gaussianWeights[distributionIndex] = gaussLikelihood * prior
        laplacianWeights[distributionIndex] = laplacianLikelihood
        
    #JITTER BEFORE YOU NORMALIZE ??   
    allGaussianWeights.append(gaussianWeights)
    allLaplacianWeights.append(laplacianWeights)
    
        
    #Normalize Weights for Sampling    
    gaussNormed = gaussianWeights / np.sum(gaussianWeights)
    laplacianNormed = laplacianWeights / np.sum(laplacianWeights)

    #Create New Proposal
    proposalIndecies = np.random.multinomial(n, gaussNormed, size=1)
    proposalValues = np.zeros((3,n), dtype=np.float)
    previousWeights = np.zeros((n))

    counter = 0
    #print len(proposalIndecies[0])
    #print proposalIndecies[0][i]
    
    for i in range(0,len(proposalIndecies[0]),1):
        for k in range(0,proposalIndecies[0][i],1):
            proposalValues[:,counter] = particles[len(particles) -1][:,proposalIndecies[0][i]]
            previousWeights[counter] = gaussianWeights[i]
            counter +=1
            
    allPrevoiusWeights.append(previousWeights)
    
    #Jitter Proposal
    for i in range (0,len(proposalValues[0]),1):
        proposalValues[:,i][0] =  (proposalValues[:,i][0] + random.uniform(-.01, .01))
        proposalValues[:,i][1] =  (proposalValues[:,i][1] + random.uniform(-.01, .01))
        proposalValues[:,i][2] =  (proposalValues[:,i][2] + random.uniform(-.5, .5))
            
    #Created Expected X and Y
    expectedX = np.mean(proposalValues[0])
    expectedY = np.mean(proposalValues[1])
    predictedXs.append(expectedX)
    predictedYs.append(expectedY)
    
    #Add particles for next time step
    if ((t + 1) != len(sonarsData[0])):
        particles.append(proposalValues)
```

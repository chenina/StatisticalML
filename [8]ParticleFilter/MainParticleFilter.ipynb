```python
import scipy.io
import numpy 
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import random
import numpy as np
import seaborn
import time
import math
from lab8_common import *
from lab8_gen_sonars import *
matlabData = scipy.io.loadmat('sensors.mat')
sonars = matlabData['sonars']
true_states = matlabData['true_states']
```

```python
def GaussianProb(parts,sonars,room_map):
    ray = cast_rays(parts,room_map)
    rayDiffs = ray - list(sonars)
    returnVal = [np.exp((-15)*np.dot(i,i)) for i in rayDiffs]
    return returnVal/sum(returnVal)

def LaplacianProb(parts,sonars,room_map):
    ray = cast_rays(parts,room_map)
    rayDiffs = ray - list(sonars)
    returnVal = [np.exp((-15)*sum(abs(i))) for i in rayDiffs]
    return returnVal/sum(returnVal)
```

```python
def resample(parts, weights):
    rows,cols = np.shape(parts)
    
    distribution = np.random.choice(np.arange(cols), p=weights,size=cols)
    proposalDistribution = np.array([parts[:,i] for i in distribution]).T
    
    proposalDistribution[0] =  [proposalDistribution[0,c] + .1*np.random.rand()-.05 for c in xrange(cols)]
    proposalDistribution[1] =  [proposalDistribution[1,c] + .1*np.random.rand()-.05 for c in xrange(cols)]
    proposalDistribution[2] =  [proposalDistribution[2,c] + 11*np.random.rand()%(2*np.pi) for c in xrange(cols)]
    
    return proposalDistribution
```

```python
def particleFilter(N=1000,verbose=True, iters=225,probabilityType='laplacian',plotFinalPath=True):
    parts = np.random.rand(3,N)
    room_map = create_map()
    
    paraticle_slides = []
    for t in xrange(0,iters,1):
        print '\r', t, '/', iters,
        weightsT = None
        if (probabilityType == 'gaussian'):
            weightsT = GaussianProb(parts,sonars[:,t],room_map)
        else:
            weightsT = LaplacianProb(parts,sonars[:,t],room_map) #Laplacian Default
        parts = resample(parts,weightsT)
        paraticle_slides.append(parts)
        #Expected Value
    
    particleMeans = np.mean(paraticle_slides,axis=2)
    
    if verbose:
        %matplotlib qt
        for t in range (0,iters,1):
            plt.figure( 1 )
            plt.clf()
            plt.scatter(true_states[0,:t],true_states[1,:t]) #plot robot path
            plt.scatter(paraticle_slides[t][0],paraticle_slides[t][1], color= 'r')
            plt.scatter(particleMeans[:t,0],particleMeans[:t,1], color = 'purple')
            show_map( room_map )
            plt.pause(.0001)
            
        plt.close()
    
    if plotFinalPath:
        %matplotlib inline
        plt.figure( 1 )
        plt.scatter(true_states[0,:t],true_states[1,:t]) #plot robot path
        plt.plot(particleMeans[:,0],particleMeans[:,1],color = 'r')
        show_map( room_map )
        plt.show()


    
```

```python
#Laplacian
particleFilter(N=1000, verbose=False) #Set Verbose to True to see Video
```

```python
#Gaussian
particleFilter(N=10000, verbose=False, probabilityType='gaussian') #Set Verbose to True to see Video
```
